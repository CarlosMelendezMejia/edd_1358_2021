```
class NodoDoble:
    def __init__(self, value, previo=None, siguiente=None):
        self.data = value
        self.prev = previo
        self.next = siguiente


class DoubleLinkedList:
    def __init__(self):
        self.__head = None
        self.__tail = None
        self.__size = 0

    def get_tail_data(self):
        return self.__tail.data

    def get_size(self):
        return self.__size

    def is_empty(self):
        return self.__size == 0

    def append(self, value):
        if self.is_empty():
            nuevo = NodoDoble(value)
            self.__head = nuevo
            self.__tail = nuevo
        else:
            nuevo = NodoDoble(value, self.__tail)
            self.__tail.next = nuevo
            self.__tail = nuevo
        self.__size += 1

    def transversal(self):
        curr_node = self.__head
        while curr_node != None:
            print("<---", curr_node.data, "---->", end="")
            curr_node = curr_node.next
        print("")

    def reverse_transversal(self):
        curr_node = self.__tail
        while curr_node != None:
            print("<---", curr_node.data, "---->", end="")
            curr_node = curr_node.prev
        print("")

    def find_from_head(self, value):
        curr_node = self.__head
        cont = 0
        while curr_node != None:
            cont += 1
            if curr_node.data == value:
                break
            curr_node = curr_node.next
        return cont

    def find_from_tail(self, value):
        curr_node = self.__tail
        cont = 0
        while curr_node != None:
            cont += 1
            if curr_node.data == value:
                break
            curr_node = curr_node.prev
        return cont

    def remove_from_head(self, value):
        curr_node = self.__head
        while curr_node != None:
            if self.__head.data == value:
                self.__head = self.__head.next
                self.__head.prev = None
                break
            if curr_node.next != None and curr_node.next.data == value:
                anterior = curr_node
                anterior.next = curr_node.next.next
                anterior.next.prev = anterior
                break
            curr_node = curr_node.next

    def remove_from_tail(self, value):
        curr_node = self.__tail

        while curr_node != None:
            if self.__tail.data == value:
                self.__tail = self.__tail.prev
                self.__tail.next = None
            if curr_node.prev != None and curr_node.prev.data == value:
                anterior = curr_node
                anterior.prev = curr_node.prev.prev
                anterior.prev.next = anterior
                break
            curr_node = curr_node.prev


    def print_prev_and_next(self):
        curr_node = self.__head
        if self.__head != None and self.__tail != None:
            print(f"Cabeza: {self.__head.data} // Cola: {self.__tail.data}")
        else:
            print("La lista esta vacia")
        while curr_node != None:
            if curr_node.prev == None:
                print(f"Nodo {curr_node.data}, anterior None, siguiente {curr_node.next.data}")
            elif curr_node.next == None:
                print(f"Nodo {curr_node.data}, anterior {curr_node.prev.data}, next None")
            else:
                print(f"Nodo {curr_node.data}, anterior {curr_node.prev.data}, siguiente {curr_node.next.data}")

            curr_node = curr_node.next
        print("")
```

```
ldl = DoubleLinkedList()

#Lista vacia
print(f"el tama単o de la lista es de {ldl.get_size()}")
print("")

#Metodo append
for i in range(10, 90, 10):
    ldl.append(i)
print(f"el tama単o de la lista es de {ldl.get_size()}")
print("")

#Metodo transversal
ldl.transversal()
print("")

#Metodo transversal reverso
ldl.reverse_transversal()
print("")

#Metodo imprimir nodos
ldl.print_prev_and_next()

#Metodo find_from_head
print(f"la posicion del 20 desde la cabeza es la {ldl.find_from_head(20)}a")
print("")

#Metodo find_from_tail
print(f"la posicion del 20 desde la cola es la {ldl.find_from_tail(20)}a")
print("")

#Metodo remove from head
ldl.remove_from_head(10)
ldl.transversal()
ldl.print_prev_and_next()

ldl.remove_from_head(40)
ldl.transversal()
ldl.print_prev_and_next()

#Metodo remove from tail
ldl.remove_from_tail(80)
ldl.transversal()
ldl.print_prev_and_next()

ldl.remove_from_tail(50)
ldl.transversal()
ldl.print_prev_and_next()
```
