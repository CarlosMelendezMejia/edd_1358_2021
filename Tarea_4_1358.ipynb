```
class nodo:

    def __init__(self, dato, siguiente=None):  # Argumento con un valor asignado
        self.dato = dato
        self.siguiente = siguiente


class Linkedlist:
    def __init__(self):
        self.__head = None

    def is_empty(self):
        return self.__head == None

    def tail(self):
        curr_node = self.__head
        while curr_node != None and curr_node.siguiente != None:
            curr_node = curr_node.siguiente
        return curr_node

    def append(self, value):
        nuevo = nodo(value)
        if self.__head == None:
            self.__head = nuevo
        else:
            self.tail().siguiente = nuevo

    def transversal(self):
        curr_node = self.__head
        if curr_node == None:
            print("Lista vacia")
        else:
            while curr_node != None:
                print(f" {curr_node.dato} ---->", end="")
                curr_node = curr_node.siguiente
        print("")

    def prepend(self, value):
        nuevo = nodo(value, self.__head)
        self.__head = nuevo

    def remove(self, valor):
        curr_node = self.__head
        while curr_node != None:
            if self.__head.dato == valor:
                self.__head = self.__head.siguiente
                break
            if curr_node.siguiente != None and curr_node.siguiente.dato == valor:
                anterior = curr_node
                anterior.siguiente = curr_node.siguiente.siguiente
                break
            curr_node = curr_node.siguiente

    def pop(self, pos=None):
        curr_node = self.__head
        cont = 0
        while curr_node != None:
            cont += 1
            if pos != None:
                if pos - 1 == cont:
                    anterior = curr_node
                    anterior.siguiente = curr_node.siguiente.siguiente
                    break
                elif pos == 1:
                    self.__head = self.__head.siguiente
                    break
            if pos == None:
                if curr_node.siguiente == None:
                    self.remove(curr_node.dato)
            curr_node = curr_node.siguiente

    def len(self):
        curr_node = self.__head
        cont = 0
        while curr_node != None:
            cont += 1
            curr_node = curr_node.siguiente
        return cont

l = Linkedlist()
print(f"L esta vacia? = {l.is_empty()}")
x = l.tail()
print(x)

# metodo append
print("Metodo append")
l.append(10)
l.transversal()
l.append(20)
l.transversal()
l.append(30)
l.transversal()
l.append(40)
l.transversal()
print("")

# metodo prepend
print("Metodo prepend")
l.prepend(5)
l.prepend(1)
l.transversal()
print("")

# Metodo remove
print("Metodo remove")
l.remove(10)
l.transversal()
print("")

# Metodo pop
print("Metodo pop vacio")
l.pop()
l.transversal()
print("Metodo pop con posicion")
l.pop(2)
l.transversal()
```
