```
class Metodos_numericos:
    def __init__(self):
        self.__numeros = 0

    def sumar(self, num1, num2):
        return num1 + num2

    def restar(self, num1, num2):
        return num1 - num2

    def multiplicar(self, num1, num2):
        return num1 * num2

    def potencia(self, num1, num2):
        return num1 ** num2

    def dividir(self, num1, num2):
        try:
            return num1 / num2
        except:
            return 0

    def separarCadena(self, cadena, numero):
        temp = ""
        sep = []
        for i in range(len(cadena)):
            if cadena[i] == "-":
                sep.append(temp)
                temp = "-"
                sep.append(temp)
                temp = ""
            elif cadena[i] == "+":
                sep.append(temp)
                temp = "+"
                sep.append(temp)
                temp = ""
            elif cadena[i] == "*":
                sep.append(temp)
                temp = "*"
                sep.append(temp)
                temp = ""
            elif cadena[i] == "/":
                sep.append(temp)
                temp = "/"
                sep.append(temp)
                temp = ""
            elif cadena[i] == "^":
                sep.append(temp)
                temp = "^"
                sep.append(temp)
                temp = ""
            elif cadena[i] == "x":
                if i != 0:
                    if cadena[i - 1] in ["*", "+", "-", "^", "/"]:
                        sep.append(temp)
                        temp = str(numero)
                        sep.append(temp)
                        temp = ""
                    else:
                        sep.append(temp)
                        temp = "*"
                        sep.append(temp)
                        temp = str(numero)
                        sep.append(temp)
                        temp = ""
                else:
                    sep.append(temp)
                    temp = str(numero)
                    sep.append(temp)
                    temp = ""
            else:
                temp += cadena[i]
        sep.append(temp)
        ##Limpiar la cadena
        for i in sep:
            if i == "":
                sep.remove(i)
        ##Ley de signos
        for i in range(len(sep)):
            if sep[i] == "-" and sep[i - 1] == "*":
                sep.remove(cadena[i])
                sep.insert(i, "-1")
                sep.insert(i + 1, "*")
            if i == 0 and sep[i] == "-":
                sep.remove(cadena[i])
                sep.insert(i, "-1")
                sep.insert(i + 1, "*")

        return sep

    def separarCadena_sin_evaluar(self, cadena):
        temp = ""
        sep = []
        for i in range(len(cadena)):
            if cadena[i] == "-":
                sep.append(temp)
                temp = "-"
                sep.append(temp)
                temp = ""
            elif cadena[i] == "+":
                sep.append(temp)
                temp = "+"
                sep.append(temp)
                temp = ""
            elif cadena[i] == "*":
                sep.append(temp)
                temp = "*"
                sep.append(temp)
                temp = ""
            elif cadena[i] == "/":
                sep.append(temp)
                temp = "/"
                sep.append(temp)
                temp = ""
            elif cadena[i] == "^":
                sep.append(temp)
                temp = "^"
                sep.append(temp)
                temp = ""
            elif cadena[i] == "x":
                sep.append(temp)
                temp = "x"
                sep.append(temp)
                temp = ""
            else:
                temp += cadena[i]
        sep.append(temp)
        ##Limpiar la cadena
        for i in sep:
            if i == "":
                sep.remove(i)
        ##Ley de signos
        for i in range(len(sep)):
            if sep[i] == "-" and sep[i - 1] == "*":
                sep.remove(cadena[i])
                sep.insert(i, "-1")
                sep.insert(i + 1, "*")
            if i == 0 and sep[i] == "-":
                sep.remove(cadena[i])
                sep.insert(i, "-1")
                sep.insert(i + 1, "*")
        return sep

    def verificarVariable(self, lista, variable):
        verificacion = False
        for i in lista:
            if i == variable:
                verificacion = True
        return verificacion

    def verificarCadena(self, lista):
        fallo = None
        signos = ['+', '-', '*', '^', '/']
        for i in lista:
            if i not in signos:
                try:
                    int(i)
                except:
                    try:
                        float(i)
                    except:
                        fallo = True

        for i in range(len(lista)):
            if i == len(lista) - 1 and lista[i] in signos:
                fallo = True

        for i in range(len(lista)):
            if i != len(lista) - 1:
                if lista[i] in signos and lista[i + 1] in signos:
                    fallo = True

        if fallo:
            valido = False
        else:
            valido = True
        return valido

    def verificarArgumentos(self, *args):
        valido = True
        for arg in args:
            try:
                int(arg)
            except:
                try:
                    float(arg)
                except:
                    valido = False
        return valido

    def calcular(self, cadena, numero):
        sep = self.separarCadena(cadena, numero)

        if self.verificarCadena(sep):
            contpor = 0
            contdiv = 0
            contres = 0
            contsum = 0
            contpot = 0
            contnum = 0

            for i in sep:
                if i == "*":
                    contpor += 1
                elif i == "^":
                    contpot += 1
                elif i == "/":
                    contdiv += 1
                elif i == "-":
                    contres += 1
                elif i == "+":
                    contsum += 1
                else:
                    contnum += 1

            while len(sep) > 1:
                while contpot > 0:
                    for i in range(len(sep)):
                        try:
                            if sep[i] == "^":
                                num1 = float(sep[i - 1])
                                num2 = float(sep[i + 1])
                                resultado = self.potencia(num1, num2)
                                sep.pop(i + 1)
                                sep.pop(i)
                                sep.pop(i - 1)
                                sep.insert(i - 1, str(resultado))
                                contpot = contpot - 1

                        except:
                            pass
                while contpor > 0:
                    for i in range(len(sep)):
                        try:
                            if sep[i] == "*":
                                num1 = float(sep[i - 1])
                                num2 = float(sep[i + 1])
                                resultado = self.multiplicar(num1, num2)
                                sep.pop(i + 1)
                                sep.pop(i)
                                sep.pop(i - 1)
                                sep.insert(i - 1, str(resultado))
                                contpor = contpor - 1

                        except:
                            pass
                while contdiv > 0:
                    for i in range(len(sep)):
                        try:
                            if sep[i] == "/":
                                num1 = float(sep[i - 1])
                                num2 = float(sep[i + 1])
                                resultado = self.dividir(num1, num2)
                                sep.pop(i + 1)
                                sep.pop(i)
                                sep.pop(i - 1)
                                sep.insert(i - 1, str(resultado))
                                contpot = contpot - 1

                        except:
                            pass
                while contres > 0:
                    for i in range(len(sep)):
                        try:
                            if sep[i] == "-":
                                num1 = float(sep[i - 1])
                                num2 = float(sep[i + 1])
                                resultado = self.restar(num1, num2)
                                sep.pop(i + 1)
                                sep.pop(i)
                                sep.pop(i - 1)
                                sep.insert(i - 1, str(resultado))
                                contres = contres - 1

                        except:
                            pass
                while contsum > 0:
                    for i in range(len(sep)):
                        try:
                            if sep[i] == "+":
                                num1 = float(sep[i - 1])
                                num2 = float(sep[i + 1])
                                resultado = self.sumar(num1, num2)
                                sep.pop(i + 1)
                                sep.pop(i)
                                sep.pop(i - 1)
                                sep.insert(i - 1, str(resultado))
                                contsum = contsum - 1

                        except:
                            pass

            return float(sep[0])
        else:
            return float(0)

    def biseccion(self, cadena, numero1, numero2, iteraciones):

        if self.verificarArgumentos(numero1,numero2,iteraciones):
            if self.verificarCadena(self.separarCadena(cadena, numero1)) and self.verificarCadena(
                    self.separarCadena(cadena, numero2)):
                # Asignar a y b iniciales
                a = {"numero": numero1, "evaluacion": self.calcular(cadena, numero1), "signo": ""}
                b = {"numero": numero2, "evaluacion": self.calcular(cadena, numero2), "signo": ""}
                # Evaluar signos en a y b
                if a["evaluacion"] < 0:
                    a["signo"] = "-"
                else:
                    a["signo"] = "+"

                if b["evaluacion"] < 0:
                    b["signo"] = "-"
                else:
                    b["signo"] = "+"
                if a["signo"] != b["signo"]:
                    for i in range(iteraciones):
                        numero3 = (a["numero"] + b["numero"]) / 2
                        c = {"numero": numero3, "evaluacion": self.calcular(cadena, numero3), "signo": ""}
                        if c["evaluacion"] < 0:
                            c["signo"] = "-"
                        else:
                            c["signo"] = "+"
                        if c["signo"] == a["signo"]:
                            a = {"numero": numero3, "evaluacion": self.calcular(cadena, numero3), "signo": ""}
                            if a["evaluacion"] < 0:
                                a["signo"] = "-"
                            else:
                                a["signo"] = "+"
                        elif c["signo"] == b["signo"]:
                            b = {"numero": numero3, "evaluacion": self.calcular(cadena, numero3), "signo": ""}
                            if b["evaluacion"] < 0:
                                b["signo"] = "-"
                            else:
                                b["signo"] = "+"
                    return {"iter": iteraciones, "a": a["evaluacion"], "b": b["evaluacion"], "raiz": c["numero"],
                            "evaluacion": c["evaluacion"], "error": abs(((a["evaluacion"] - b["evaluacion"]) / 2) * 100)}
                else:
                    return "La funcion no cruza el eje X en los valores dados"
            else:
                return "La función no es valida"
        else:
            return "Los argumentos no son validos"

    def newton_rapshon(self, valor, fun, der, iter):

        if self.verificarCadena(self.separarCadena(fun, valor)) and self.verificarCadena(
                self.separarCadena(der, valor)):
            Xinicial = valor
            division = self.calcular(fun, Xinicial) / self.calcular(der, Xinicial)
            X = valor - division

            if iter > 1:
                for i in range(iter + 1):
                    anterior = X
                    Xinicial = X
                    division = self.calcular(fun, Xinicial) / self.calcular(der, Xinicial)
                    X = valor - division
                    error = abs(X - anterior)
                return {"iter": i, "Xi": Xinicial, "Xi+1": X, "error": error}
            else:
                return {"iter": 1, "Xi": Xinicial, "Xi+1": X}
                # return "1a iteracion. X inicial: " + str(Xinicial) + " Xi+1: " + str(X)
        else:
            return "La función o la derivada no es valida"

    def gauss_seidel(self, x1, y1, z1, r1, x2, y2, z2, r2, x3, y3, z3, r3, iter, x0, y0, z0):
        Ec1 = [x1, y1, z1, r1]
        Ec2 = [x2, y2, z2, r2]
        Ec3 = [x3, y3, z3, r3]
        val = [Ec1, Ec2, Ec3, [iter, x0, y0, z0]]
        valido = True
        for i in val:
            for j in i:
                try:
                    int(j)
                except:
                    try:
                        float(j)
                    except:
                        valido = False

        if valido:
            valorX = Ec1[3] - Ec1[1] * y0 - Ec1[2] * z0
            valorX = valorX / Ec1[0]
            valorY = Ec2[3] - Ec2[0] * valorX - Ec2[2] * z0
            valorY = valorY / Ec2[1]
            valorZ = Ec3[3] - Ec3[0] * valorX - Ec3[1] * valorY
            valorZ = valorZ / Ec3[2]
            if iter == 1:
                return f"x: {valorX} y: {valorY} z:{valorZ}"
            else:
                for i in range(iter - 1):
                    Xant = valorX
                    Yant = valorY
                    Zant = valorZ
                    valorX = Ec1[3] - Ec1[1] * Yant - Ec1[2] * Zant
                    valorX = valorX / Ec1[0]
                    valorY = Ec2[3] - Ec2[0] * valorX - Ec2[2] * Zant
                    valorY = valorY / Ec2[1]
                    valorZ = Ec3[3] - Ec3[0] * valorX - Ec3[1] * valorY
                    valorZ = valorZ / Ec3[2]
                    error_x = valorX - Xant
                    error_x = abs(error_x / valorX) * 100
                    error_y = valorY - Yant
                    error_y = abs(error_y / valorY) * 100
                    error_z = valorZ - Zant
                    error_z = abs(error_z / valorZ) * 100
                return {"iter": iter, "x": valorX, "%error x": error_x, "y": valorY, "%error y": error_y, "z": valorZ,
                        "%error z": error_z}
                # return f"iteracion: {iter} x: {valorX} error_X: {error_x}% y: {valorY} error_Y: {error_y}% z:{valorZ} error_Z: {error_z}%"
        else:
            return "Introduzca solamente numeros"

    def simpson_un_tercio(self, fun, limA, limB, seg):
        if self.verificarArgumentos(limA, limB, seg):
            b = limB
            a = limA
            h = b - a
            h = h / seg
            h_entre_3 = h / 3
            if self.verificarCadena(self.separarCadena(fun, a + 1 * h)):
                Xi = []
                fx = []
                pares = []
                impares = []
                resultado = []
                par = 0
                impar = 0
                final = 0
                if seg % 2 == 0:
                    for n in range(seg + 1):
                        Xi.append(a + n * h)
                    for i in range(len(Xi)):
                        fx.append(self.calcular(fun, Xi[i]))

                    for i in range(len(fx)):
                        if i == 0 or i == len(fx) - 1:
                            resultado.append(fx[i])
                        elif i != 0 or i != len(fx) - 1:
                            if i % 2 == 0:
                                pares.append(fx[i])
                            else:
                                impares.append(fx[i])
                    for i in pares:
                        par += i
                    for i in impares:
                        impar += i
                    par = par * 2
                    impar = impar * 4
                    resultado.append(par)
                    resultado.append(impar)
                    for i in resultado:
                        final += i
                    final = final * h_entre_3
                    return {"fun": fun, "resultado": final}
                    # return f"El resultado de la integral por simpson 1/3 de {fun} es de {final} u^2"
                else:
                    return "El numero de segmentos no es par"
            else:
                return "La función no es valida"
        else:
            return "El limite o los segmentos no son validos, introduce numeros"

    def simpson_tres_octavos(self, fun, limA, limB, seg):
        if self.verificarArgumentos(limA, limB, seg):
            b = limB
            a = limA
            h = b - a
            h = h / seg
            h_tres_octavos = h * 3
            h_tres_octavos = h_tres_octavos / 8

            if self.verificarCadena(self.separarCadena(fun, a + 1 * h)):
                Xi = []
                fx = []
                resultado = []
                final = 0
                if seg == 3:
                    for n in range(seg + 1):
                        Xi.append(a + n * h)
                    for i in range(len(Xi)):
                        fx.append(self.calcular(fun, Xi[i]))
                    for i in range(len(fx)):
                        if i == 0 or i == 3:
                            resultado.append(fx[i])
                        elif i == 1 or i == 2:
                            resultado.append(fx[i] * 3)
                    for i in resultado:
                        final += i
                    final = final * h_tres_octavos

                    return {"fun": fun, "resultado": final}

                else:
                    return "El numero de segmentos debe ser de 3"
            else:
                return "La función no es valida"
        else:
            return "El limite o los segmentos no son validos, introduce numeros"

    def simpson_mixto(self, fun, limA, limB, seg):
        if self.verificarArgumentos(limA,limB,seg):
            b = limB
            a = limA
            h = b - a
            h = h / seg
            h_tres_octavos = h * 3
            h_tres_octavos = h_tres_octavos / 8
            h_entre_3 = h / 3
            if self.verificarCadena(self.separarCadena(fun,a+1+h)):
                Xi = []
                fx = []
                resultado = []
                res_3_8 = 0
                res_1_3 = 0
                pares = []
                impares = []
                par = 0
                impar = 0
                if seg % 2 != 0:
                    for n in range(seg + 1):
                        Xi.append(a + n * h)
                    for i in range(len(Xi)):
                        fx.append(self.calcular(fun, Xi[i]))
                    tres_octavos = fx[0:4]
                    un_tercio = fx[3:len(fx)]

                    ####Tres octavos

                    for i in range(len(tres_octavos)):
                        if i == 0 or i == 3:
                            resultado.append(tres_octavos[i])
                        elif i == 1 or i == 2:
                            resultado.append(tres_octavos[i] * 3)
                    for i in resultado:
                        res_3_8 += i
                    res_3_8 = res_3_8 * h_tres_octavos
                    resultado = []

                    #####Un tercio

                    for i in range(len(un_tercio)):
                        if i == 0 or i == len(un_tercio) - 1:
                            resultado.append(un_tercio[i])
                        elif i != 0 or i != len(un_tercio) - 1:
                            if i % 2 == 0:
                                pares.append(un_tercio[i])
                            else:
                                impares.append(un_tercio[i])
                    for i in pares:
                        par += i
                    for i in impares:
                        impar += i
                    par = par * 2
                    impar = impar * 4
                    resultado.append(par)
                    resultado.append(impar)
                    for i in resultado:
                        res_1_3 += i
                    res_1_3 = res_1_3 * h_entre_3

                    final = res_1_3 + res_3_8
                    return {"fun": fun, "resultado": final}
                else:
                    return "El segmento debe ser impar, se recomienda usar simpson 1/3"
            else:
                return "La función no es valida"
        else:
            return "Los argumentos no son validos"


m = Metodos_numericos()

print(m.biseccion("x^4+3x^3-2", 0, 1, 6))
print(m.newton_rapshon(-2.5, "x^3-3x+4", "3x^2-3", 4))
print(m.gauss_seidel(9, 2, -1, -2, 7, 8, 5, 3, 3, 4, -10, 6, 6, 0, 0, 0))
print(m.simpson_un_tercio("x^3+2x^2+5x", 0, 1, 6))
print(m.simpson_tres_octavos("x^3+2x^2+5x", 0, 1, 3))
print(m.simpson_mixto("x^3+2x^2+5x", 0, 1, 9))
```
