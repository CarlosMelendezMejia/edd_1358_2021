```
lista = [4,5,2,1,9,8,7,6,3]
lista_1a3 = []
lista_4a6 = []
lista_7a9 = []
for i in range(0,len(lista)):
  if lista[i] >0 and lista[i]<=3:
    lista_1a3.append(lista[i])
  elif lista[i] >3 and lista[i]<=6:
    lista_4a6.append(lista[i])
  elif lista[i] >6 and lista[i]<=9:
    lista_7a9.append(lista[i])

while lista_1a3!=[1,2,3]:
  for i in range(0,len(lista_1a3)):
    if i==len(lista_1a3)-1:
      print("entro if")
      num1=lista_1a3[i]
      num2=lista_1a3[i-1]
      if num1<num2:
        print("entro if 2")
        lista_1a3[i]=num2
        lista_1a3[i-1]=num1
    elif i==0 or i==1:
      print("entro elif")
      num1=lista_1a3[i]
      num2=lista_1a3[i+1]
      if num1<num2:
        print("entro if de elif")
        lista_1a3[i]=num1
        lista_1a3[i+1]=num2
      else:
        print("entro else de elif")
        lista_1a3[i]=num2
        lista_1a3[i+1]=num1
print(lista_1a3)

while lista_4a6!=[4,5,6]:
  for i in range(0,len(lista_4a6)):
    if i==len(lista_4a6)-1:
      print("entro if")
      num1=lista_4a6[i]
      num2=lista_4a6[i-1]
      if num1<num2:
        print("entro if 2")
        lista_4a6[i]=num2
        lista_4a6[i-1]=num1
    elif i==0 or i==1:
      print("entro elif")
      num1=lista_4a6[i]
      num2=lista_4a6[i+1]
      if num1<num2:
        print("entro if de elif")
        lista_4a6[i]=num1
        lista_4a6[i+1]=num2
      else:
        print("entro else de elif")
        lista_4a6[i]=num2
        lista_4a6[i+1]=num1
print(lista_4a6)

while lista_7a9!=[7,8,9]:
  for i in range(0,len(lista_7a9)):
    if i==len(lista_7a9)-1:
      print("entro if")
      num1=lista_7a9[i]
      num2=lista_7a9[i-1]
      if num1<num2:
        print("entro if 2")
        lista_7a9[i]=num2
        lista_7a9[i-1]=num1
    elif i==0 or i==1:
      print("entro elif")
      num1=lista_7a9[i]
      num2=lista_7a9[i+1]
      if num1<num2:
        print("entro if de elif")
        lista_7a9[i]=num1
        lista_7a9[i+1]=num2
      else:
        print("entro else de elif")
        lista_7a9[i]=num2
        lista_7a9[i+1]=num1
print(lista_7a9)
nueva_lista=[]
nueva_lista.extend(lista_1a3)
nueva_lista.extend(lista_4a6)
nueva_lista.extend(lista_7a9)

print(nueva_lista)


```
