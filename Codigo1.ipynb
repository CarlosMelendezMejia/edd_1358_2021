```
class Array2D:#Declara la clase Array2D
    def __init__(self, ren, col): #Constructor con dos argumentos (renglones, columnas)
        self.renglon = ren
        self.columna = col
        self.arreglo = [[0 for i in range(col)] for i in range(ren)] #Generador, crea una lista de 2 dimensiones en base a las medidas establecidas

    def clear(self, dato):
        self.arreglo = [[[dato] for i in range(self.columna)] for i in range(self.renglon)] #Llena el array con un dato en especifico o lo limpia

    def get_row_size(self): #Regresa el tamaño de la fila
        return self.renglon

    def get_col_size(self): #Regresa el tamaño de la columna
        return self.columna

    def set_item(self, ren, col, dato):  #Establece un onjeto especificado en la posicion indicada mediante el renglon y la columna
        self.arreglo[ren][col] = dato

    def get_item(self, ren, col): #Regresa un onjeto especificado en la posicion indicada mediante el renglon y la columna
        return self.arreglo[ren][col]

    def to_string(self): #Metodo to String
        return self.arreglo


class JuegoDeLaVida:

    def __init__(self, rens, cols, generaciones, poblacion): 
        self.largo = cols
        self.alto = rens
        self.grid = Array2D(self.alto, self.largo) #Inicias el objeto de la clase Array2d
        self.generaciones = generaciones

        for i in poblacion:
            self.grid.set_item(i[0], i[1], 1) #En base a las coordenadas de poblacion, establece las celulas vivas, las cuales son representadas con un 1

    def imprime_grid(self):
        for r in range(self.grid.get_col_size()):
            for c in range(self.grid.get_row_size()):
                if self.grid.get_item(r, c) == 0:

                    print('░░', end='')
                else:
                    print('██', end='')

            print('')
        print('')

    def get_num_rows(self):  # Nos da el número de filas
        return self.alto

    def get_num_cols(self):  # Nos da el número de columnas
        return self.largo

    def set_celula_muerta(self, row, col):  # Hace que una celula esté muerta, con el valor 0
        self.grid.set_item(row, col, 0)

    def set_celula_viva(self, row, col):  # Hace que una celula esté viva, con el valor 1
        self.grid.set_item(row, col, 1)

    def get_is_viva(self, row, col):  # Regresa true si en el lugar seleccionado hay una celula viva
        return self.grid.get_item(row, col) == 1

    def get_is_muerta(self, row, col):  # Regresa true si en el lugar seleccionado hay una celula viva
        return self.grid.get_item(row, col) == 0

    def get_numero_vecinos_vivos(self, row, col):  # Nos dice cuantos vecinos vivos tiene una celda
        CONTADOR_VECINOS = 0
        #Revisa las 8 celdas alrededor de las celulas, en caso de haber menos de 8, entre el try y hace una excepcion
        for i in [row - 1, row, row + 1]:
            for j in [col - 1, col, col + 1]:
                try: 
                    if self.grid.get_item(i, j) == 1 and (i, j) != (row, col):
                        CONTADOR_VECINOS += 1 #Suma las celulas vivas
                except Exception as e:
                    pass

        return CONTADOR_VECINOS

    def evolucionar(self):  # Aqui ocurren los cambios de generacion a generacion

        for i in range(self.generaciones): #todo dentro de un for


            self.imprime_grid() 

            lista_proximo_vivos = [] #Debes guardar las coordenadas de los vivos y los muertos en 2 listas que no se modifiquen
            lista_proximo_muertos = []

            for row in range(self.alto):
                for col in range(self.largo):
                  #Reglas
                    if self.get_numero_vecinos_vivos(row, col) == 3: #Si hay 3 vecinos alrededor de la celula
                        lista_proximo_vivos.append((row, col)) #Esta vive y se guarda en la lista

                    if self.get_numero_vecinos_vivos(row, col) == 2: #Si hay 2 vecinos alrededor de la celula
                        pass #Esta vive y se guarda en la lista

                    if self.get_numero_vecinos_vivos(row, col) < 2 or self.get_numero_vecinos_vivos(row, col) > 3: #Si hay menos de 2 o mas de 3 vecinos
                        lista_proximo_muertos.append((row, col)) #Esta muere y se guarda en la lista

            for i in lista_proximo_vivos:
                self.set_celula_viva(i[0], i[1]) #Se establecen primero los vivos en la tabla

            for i in lista_proximo_muertos:
                self.set_celula_muerta(i[0], i[1]) #Se establecen despues los muertos en la tabla


x = JuegoDeLaVida(7, 7, 7, [(1, 2), (2, 2), (2, 1), (2, 3)])  # Ejemplo, se crea el objeto juego de la vida, el ultimo parametro es una lista de tuplas

x.evolucionar()#Se llama al metodo evolucionar
```
