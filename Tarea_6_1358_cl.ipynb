```
class Nodo:
    def __init__(self, value, siguiente=None):
        self.data = value
        self.next = siguiente

class CircularList:
    def __init__(self):
        self.__ref = None
        self.__size = 0

    def is_empty(self, ):
        return self.__size == 0

    def get_size(self):
        return self.__size

    def insert(self, value):
        curr_node = self.__ref
        if curr_node == None:
            nuevo = Nodo(value)
            self.__ref = nuevo
            nuevo.next = nuevo
            self.__size += 1
        elif value > curr_node.data:
            anterior = curr_node.next
            nuevo = Nodo(value)
            curr_node.next = nuevo
            curr_node.next.next = anterior
            self.__ref = curr_node.next
            self.__size += 1
        elif value < curr_node.data:
            if value < curr_node.next.data:
                nuevo = Nodo(value)
                siguiente = curr_node.next
                curr_node.next = nuevo
                curr_node.next.next = siguiente
                self.__size += 1
            else:
                nuevo = Nodo(value)
                curr_node = self.__ref.next
                while curr_node != self.__ref:
                    if curr_node.data < value and curr_node.next.data > value:
                        siguiente = curr_node.next
                        curr_node.next = nuevo
                        curr_node.next.next = siguiente
                        self.__size += 1
                        break
                    curr_node = curr_node.next

    def transversal(self):
        curr_node = self.__ref
        print(f"{curr_node.data}---->",end="")
        curr_node = self.__ref.next
        while curr_node != self.__ref:
            if curr_node.next == self.__ref:
                print(f"{curr_node.data}---->{curr_node.next.data}", end="")
            else:
                print(f"{curr_node.data}---->", end="")
            curr_node = curr_node.next
        print("")

    def search(self,value):
        curr_node = self.__ref.next
        validacion = False
        cont = self.get_size()
        while curr_node != self.__ref or cont > 0:
            if curr_node.data == value :
                validacion = True
            curr_node = curr_node.next
            cont = cont - 1
        return validacion

    def remove(self,value):
        curr_node = self.__ref
        salida = True
        if self.search(value):
            while salida:
                if curr_node.next.data == value:
                    if value == self.__ref.data:
                        curr_node.next = curr_node.next.next
                        self.__ref = curr_node
                        salida = False
                    else:
                        curr_node.next = curr_node.next.next
                        salida = False
                curr_node = curr_node.next
        else:
            print("No se encuentra el valor en la lista")

    def print_ref_and_next(self):
        curr_node = self.__ref.next
        if self.__ref != None:
            print(f"Referencia: {self.__ref.data}, Siguiente {self.__ref.next.data}")
        while curr_node.data != self.__ref.data:
            print(f"nodo = {curr_node.data}, siguiente = {curr_node.next.data}")
            curr_node = curr_node.next
        print("")
```

```
cl = CircularList()

print(f"La lista esta vacia? --->{cl.is_empty()}")

#Metodo insert

cl.insert(10)
print(f"La lista esta vacia? --->{cl.is_empty()}")
cl.print_ref_and_next()

cl.insert(20)
cl.insert(30)
cl.insert(40)
cl.insert(50)
cl.insert(60)
cl.print_ref_and_next()

cl.insert(5)
cl.insert(45)
cl.insert(70)
cl.print_ref_and_next()

#Metodo transversal
cl.transversal()

#Metodo search
print(f"el valor 50 se encuentra en la lista? ---> {cl.search(50)}")
print(f"el valor 90 se encuentra en la lista? ---> {cl.search(90)}")
print("")

#Metodo remove
cl.remove(70)
cl.remove(20)
cl.remove(5)
cl.remove(1)
cl.transversal()
```
